```python
Bikes['profit'] = Bikes['selling price'] - Bikes['standard cost']
```

### Product Size

```python
Bikes['product_size'].unique() 
```

```python
Bikes['product_size'].value_counts(normalize=True)*100
```

```python
product_size = Bikes['product_size'].value_counts(normalize=True)*100
```

```python
product_size.index # get the index of the product size
```

```python
product_size.values
```

```python
#plotting a pie chart
plt.pie(product_size, labels=product_size.index, autopct='%1.2f%%')
plt.title('Distribution of Product Size');
```

```python
#product size profit
product_size_proft = Bikes.groupby('product_size')['profit'].sum().sort_values()
product_size_proft
```

```python
fig, ax = plt.subplots(figsize=(10,5))
bars = ax.barh(product_size_proft.index, product_size_proft, color='gold')

# list comprehension, formatting
ax.bar_label(bars, [f'{value:,.0f}' for value in product_size_proft], padding=3)
plt.title('Profit Distribution by Product Size')
plt.xticks(ticks=[])
plt.show()
```

## Brand 

```python
# frequency of brand
brand_frequency = round(Bikes['brand'].value_counts(normalize=True)*100,2)
```

```python
brand_frequency
```

```python
fig, ax = plt.subplots(figsize=(10,5))
bars = ax.bar(brand_frequency.index, brand_frequency, color='gray')

#list comprehension, formatting
ax.bar_label(bars, label= [f'{value:,.0f}' for value in brand_frequency], padding=4)
plt.title('Distribution of Brand Purchase')
plt.yticks(ticks=[])
plt.show()
```

7. The product class has to do with the quality of the product. Medium quality products has around 70% patronage, while low quality products has a patronage of a little above 15%; high quality products amazingly has a close range of patronage with high quality products. The insight I got from the data is that, customers often patronize medium or average quality products, possibly because of their purchasing power. Customers often go for what they could afford, regardless of quality, especially if it could meet their needs. Thus, majority of the customers go for medium quality products, while the rich/wealthy customers go for high quality products, while the poor ones go for low quality ones. The range of the purchase between the wealthy and the poor customers is almost the same as shown in the chart.

It also tells a lot about the customer distribution of bikers' nation, and their purchasing power. Furthermore, low quality products has a good patronage at a little bit above $1.2m. 

While high quality products received the least patronage and profit, having garnered around $1.1m. This shows that, customers prefer an average quality products to high class products; not necessarily because they don't like it, but because they can't afford it!

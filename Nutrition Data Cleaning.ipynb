```python
nutri[needed_cols].iloc[8651]
```

```python
units_a = nutri[needed_cols].iloc[8651].str.replace('[0-9]','',regex=True).str.replace('.', '', regex=False).str.strip()
units_a
```

```python
len(units_a)
```

```python
zipped = zip(nutri[needed_cols], units_a)
zipped
```

```python
type(zipped)
```

```python
new_columns =[]
for old, new in zipped:
    new_col = f'{old}_{new}'
    new_columns.append(new_col)
```

```python
type(new_columns)
```

```python
new_columns
```

```python
column_dict = {}
for i in range(len(needed_cols)):
    column_dict[needed_cols[i]] = new_columns[i]
```

```python
column_dict
```

### Renaming the columns

```python
nutri.rename(columns=column_dict, inplace=True)
nutri.head()
```

### Changing all the rows into numeric

```python
nutri[new_columns].iloc[8651].replace('[a-zA-Z]', '', regex=True).astype('float')
```

```python
for col in new_columns:
    nutri[col] = nutri[col].replace('[a-zA-Z]', '', regex=True).astype('float')
```
